```python
current_x=5            #set current value from where gradient starts
learning_rate=0.01     #set learning rate i.e. alpha
prev_step_size=1.1     #step size
precision=0.00000001
max_iters=10000        #max no. of iterations
iters=0
df=lambda x:2*(x-9)
while iters<max_iters and prev_step_size >precision :
    prev_x=current_x
    current_x=current_x-learning_rate * df(prev_x)
    prev_step_size=abs(prev_x-current_x)
print("local minimum occurs at ",current_x)
```
